# Bibliothèque matplotlib

```python
# fonction magique pour des graphes dans un notebook :
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(-np.pi, np.pi, 100) # une liste de 100 valeurs entre -pi et pi
Y = np.sin(X)
plt.plot(X, Y, color="red", linewidth=3.5, linestyle="-")
plt.plot(X, np.cos(X))
plt.savefig("graph") # sauvegarde l'image sous le nom graph.png
plt.show()           # affiche l'image à l'écran
```

# Bibliothèque calysto pour Turtle

```python
from calysto.graphics import *
from calysto.display import display, clear_output
import time
#import math

canvas = Canvas(size=(400, 400))

turtle1 = Turtle(canvas, (200, 200), 0)

for i in range(200):
    turtle1.left(88)
    turtle1.forward(1 + i)
    clear_output(wait=True)
    display(canvas)
    time.sleep(.1)
```

```python
from mobilechelonian import Turtle
t = Turtle()
t.speed(5)
colours=["red","blue","yellow","brown","black","purple","green"]

t.penup(); t.left(90); t.forward(200);t.right(90);t.pendown()
for i in range (0,18):
    t.pencolor(colours[i%7])
    t.right(20)
    t.forward(50)

t.right(180)
t.home()
```
